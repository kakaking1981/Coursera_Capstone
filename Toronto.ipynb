## Segmenting and Clustering Neighborhoods in Toronto

```python
import requests
import lxml.html as lh
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

# Create a handle, page to handle the contents of the website
page = requests.get(url)

# Store the contents of the website under doc
doc = lh.fromstring(page.content)

# Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
```

```python
# Parse Table Header
tr_elements = doc.xpath('//tr')

# Create empty list
col = []
i = 0

# For each row, store each first element {header} and an empty list
for t in tr_elements[0]:
    i = i + 1
    name = t.text_content()
    print(str(i)+':'+name)
    col.append((name,[]))
```

```python
# data from second row and after are real data
for j in range(1, len(tr_elements)):
    T = tr_elements[j]
    if len(T)!=3:
        break
    i = 0
    for t in T.iterchildren():
        data = t.text_content()
        #print(data)
        col[i][1].append(data)
        i = i + 1
             
```

```python
#print(col)
Dict = {title:column for (title, column) in col}
df = pd.DataFrame(Dict)
```

```python
df['Neighbourhood\n'].replace(r'\s+|\\n', ' ', regex=True, inplace=True)
```

```python
df
```

```python
# Remove 'Not assigned' value from column 'Borough'
df_1 = df[df['Borough']!='Not assigned']
df_1
```

```python
# (Stops here, will get back tomorrow morning) Put value of 'Not assigned' in column 'Neighbourhood' the same as value in 'Borough'
df_1.loc[df_1['Neighbourhood\n'] == 'Not assigned ','Neighbourhood\n'] = df_1.loc[df_1['Neighbourhood\n'] == 'Not assigned ','Borough']
```

```python
# Put Neighbourhood value in a list according to same Postcode
df_2 = df_1.groupby(['Postcode','Borough'])['Neighbourhood\n'].apply(list)
```
